```python
def contour_plot(f, xs, ys):
    vf = np.vectorize(f)
    X,Y = np.meshgrid(xs, ys)
    Z = vf(X,Y)
    CS = plt.contour(X, Y, Z, levels=np.linspace(np.min(Z), np.max(Z), 25))
    plt.clabel(CS, inline=1, fontsize=10)
```

```python
xs = np.linspace(-10,10,100)
ys = np.linspace(-10,10,100)
def f(x,y):
    return (x-1)**2+(y-2)**2 + (x-y)**2
contour_plot(f, xs, ys)
```
